```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from datetime import datetime
sns.set()
```

```python
# variables
year = 3
principal = 1000
amount_per_month = 100
```

```python
with open('./gold.json', 'r') as file:
    data = json.load(file)
```

```python
timestamp, value = zip(*data['chartdata'])
```

```python
df = pd.DataFrame({'timestamp':timestamp,'price':value})
```

```python
df['timestamp'] = df['timestamp'].apply(lambda x: datetime.fromtimestamp(x/1000))
```

```python
df['return'] = df['price'].pct_change()
```

```python
df['month'] = df['timestamp'].apply(lambda x: x.month)
df['day'] = df['timestamp'].apply(lambda x: x.day)
df['year'] = df['timestamp'].apply(lambda x: x.year)
```

```python
def is_the_first_time(last, new):
    if last != new:
        return True
    else:
        return False
```

```python
last_month = 0
first_of_the_month_list = []
for idx, row in df.iterrows():
    result = (last_month != row['month'])
    if result:
        first_of_the_month_list.append(1)
        last_month = row['month']
    else:
        first_of_the_month_list.append(0)
```

```python
df['first_of_the_month'] = first_of_the_month_list
```

```python
df['investments'] = df['first_of_the_month'].apply(lambda x: x*amount_per_month)
```

```python
df.plot(figsize=(10,7))
```

```python
# Calculate the compound interest and reinvesting
# So a user is investing amount per month every month for x year
```

```python
df
```

```python
# We initiate the balance
df['balance'] = np.repeat(0, len(df))
```

```python
new_balance_list = []

for idx, row in df.iterrows():
    current_row = df.iloc[idx,:]
    last_row = df.iloc[idx-1,:]
    
    if idx == 0:
        last_investment = 0
        interest_rate = 0
        last_balance = principal
    else:
        last_investment = last_row['investments']
        interest_rate = current_row['return']
        last_balance = new_balance_list[-1]
        
    new_balance = (last_balance + last_investment) * (1+interest_rate)
    new_balance_list.append(new_balance)
    print(f"last balance : {round(last_balance,2)}\tnew balance : {round(new_balance,2)}\t current return : {round(interest_rate,2)}")
```

```python
df.columns
```

```python
df['balance'] = new_balance_list
df['invested_amount'] = df['investments'].cumsum() + principal
df['amount_earned'] = df.apply(lambda x: x['balance'] - x['invested_amount'],axis=1)
```

```python
df
```

```python
df.set_index('timestamp', drop=True, inplace=True)
```

```python
df.balance.plot(figsize=(12,6), title='Evolution de la balance')
```

```python
fig, axes = plt.subplots(1,1, figsize=(12,12))
axes.fill_between(df.index, df.amount_earned + df.invested_amount, label="Montant gagnÃ©")
axes.plot(df.index, df.invested_amount, label='Montant investi', color='orange', ls='--')
axes.legend()
axes.set_title('Ã‰volution de la balance')
```
